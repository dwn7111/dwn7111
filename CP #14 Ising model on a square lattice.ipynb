```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import random
```

```python
avg=0 #to store avg.(sum(S(i))/N)
N=10 #length of lattice side 
P=np.zeros(1000) #to store probability distribution of magnetization
M=np.zeros(1000)
def repeat(N): #function to repeat process, taking the argument N, the lattice side length
    S=np.zeros((N,N)) #to store the spins of each point of the lattice
    global avgE #to store the average energy
    sumS=0 #to store the sum of S(i)s
    global avgS #to store the avg. value of a specific S(i), say S[3][3]
    for i in range(0,N):
        for j in range(0,N):
            S[i][j]=np.random.choice([-1,1]) #-1 or 1 spin assigned randomly to each point-->equal probability
            j+=1
        i+=1
    avgS+=S[3][3]
    for i in range(0,N):
        for j in range(0,N):
            sumS+=S[i][j]
    for i in range(0,N):
        for j in range(0,N):
            if j==N-1:
                if i==N-1:
                    avgE+=-((S[i][j]*S[0][j])+(S[i][j]*S[i][0]))
                else:
                    avgE+=-((S[i][j]*S[i+1][j])+(S[i][j]*S[i][0]))
            else:
                if i==N-1:
                    if j==N-1:
                        avgE+=-((S[i][j]*S[0][j])+(S[i][j]*S[i][0]))
                    else:
                        avgE+=-((S[i][j]*S[0][j])+(S[i][j]*S[i][j+1]))
                else:
                    avgE+=-((S[i][j]*S[i+1][j])+(S[i][j]*S[i][j+1]))
            j+=1
        i+=1
    m=int(sumS)
    P[m]+=1
    M[m]=sumS/N
    return sumS/N
avgS=0
avgE=0
for p in range(0,1000):
    avg+=repeat(N)
    p+=1
avg/=1000
avgS/=1000 
avgE/=1000
print("Average magnetization per spin = ",avg)
print("Average spin value of specfic point S[3][3] = ",avgS)
print("Average energy = ",avgE)
print(P)
plt.figure()
#plt.hist(P,bins=20) #redefine P array so that it can handle negative indices
plt.scatter(M,P) #close to a Gaussian distribution
#fit a Gaussian to this and find sigma^2 (variance) from this, then compare to analytically found variance
```

```python
#to make S a 1-D array instead of 2-D
#to use Monte Carlo method and see the change in E and m on changing some spins in S
arrm=np.zeros(1000) #to store each avgm for 1000 different MonteCarlo steps
def MonteCarlo(Esum,counter):
    global count #to keep track of how many MC steps occur
    global avgE,avgE2,avgm,avgm2
    N=5
    n=N*N
    T=10 #temperature
    x=np.linspace(0,N-1,N) #defining x and y co-ordinates
    y=np.linspace(0,N-1,N)
    I=np.zeros(n) 
    k=0
    Ip1=0
    Im1=0
    Iu1=0
    Id1=0
    nnR=np.zeros(n)
    nnL=np.zeros(n)
    nnU=np.zeros(n)
    nnD=np.zeros(n)
    S=np.zeros(n)
    Esum=0
    msum=0
    avgE=0
    avgm=0
    temp=0 #to hold I[k] values
    for i in range(0,N): #represents each y
        for j in range(0,N): #represents each x
            I[k]=i*N+j #I=(y-1)*N+x
            Ip1=I[k]+1
            Im1=I[k]-1
            if j==N-1:
                Ip1=I[k]-N+1
            if j==0:
                Im1=I[k]+N-1
            temp=int(I[k])
            nnR[temp]=Ip1
            nnL[temp]=Im1
            Iu1=I[k]+N
            Id1=I[k]-N
            if i==N-1:
                Iu1=j
            if i==0:
                Id1=I[k]+N*(N-1)
            nnU[temp]=Iu1
            nnD[temp]=Id1
            S[k]=np.random.choice([1]) #for all spin up 
            R=int(nnR[k])
            L=int(nnL[k])
            U=int(nnU[k])
            D=int(nnD[k])
            Esum+=-S[k]*(S[R]+S[L]+S[U]+S[D])
            msum+=S[k]
            k+=1
    for p in range(0,100):
        index=int(np.random.choice(I)) #choosing one element of I randomly to pull out one spin S[I] randomly
        delE=2*S[index]+(S[int(nnR[index])]+S[int(nnL[index])]+S[int(nnU[index])]+S[int(nnD[index])])
        print("delE=",delE)
        if delE<=0:
            S[index]=-S[index]
            Esum+=delE
            msum+=2*S[index]
        else:
            r=np.random.random() #r is a random number between 0 and 1(not including 1)
            if np.exp(-delE/T)>r:
                S[index]=-S[index]
                Esum+=delE
                msum+=2*S[index]
        print(index)
        p+=1
    Esum/=2
    print(Esum)
    print(msum)
    avgE+=Esum
    avgE2+=Esum**2
    avgm+=msum
    avgm2+=msum**2
    arrm[counter-1]=avgm
    #this is one Monte Carlo step/one timestep
    while(counter<1000 and Esum):
        counter+=1
        count+=1
        return MonteCarlo(Esum,counter)
avgE=0
avgE2=0
avgm=0
avgm2=0
count=0
MonteCarlo(-25,0)
avgm/=1000
avgm2/=1000
t=np.linspace(0,count-1,count) #array of time/Monte Carlo steps
arrm.resize(count)
plt.figure()
plt.plot(t,arrm)
plt.xlabel("Time(in s)")
plt.ylabel("Magnetization m")
```
